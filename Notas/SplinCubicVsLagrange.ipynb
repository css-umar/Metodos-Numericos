```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
from scipy.interpolate import lagrange
x = np.array([1,5,20,40])   # Presi√≥n
y = np.array([56.5,113,181,214.5])   # Temperatura
S = CubicSpline(x,y,bc_type="natural")
S
```

```python
pol = lagrange(x,y)
```

```python
xinter = np.linspace(1,40,endpoint=True)
plt.figure(figsize=(6,4))
plt.plot(x,y,"ko")
plt.plot(xinter,S(xinter),"b-")
plt.plot(xinter,pol(xinter),"k-")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
