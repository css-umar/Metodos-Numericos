```python
import numpy as np
import matplotlib.pyplot as plt
```

# Newton-Raphson

```python
def fx(x):
    return x**3+2*x**2+10*x-20
    

def dfx(x):
    return 3*x**2+4*x+10
   
```

```python
x=np.linspace(-5,10)
#datos = [0+x for x in range(20)]

x0=8.0
fx0=fx(x0)

x1=x0-(fx(x0)/dfx(x0))
fx1=fx(x1)

x2=x1-(fx(x1)/dfx(x1))
fx2=fx(x2)
```

```python
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.plot(x,fx(x))
plt.plot([x0,x1,x2], [fx0,fx1,fx2], 'o')
plt.plot([x0,x1], [fx0,0], '--k')
plt.plot([x1,x2], [fx1,0], '--k')
plt.plot([x0,x1,x2,1.36881], [0,0,0,0], 'o')


plt.plot([x0,x0], [0,fx0], '-k')
plt.plot([x1,x1], [0,fx1], '-k')


plt.annotate(r'$f(x)$', xy=(10,fx(10)), xytext=(6, 1000),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)
plt.annotate(r"$Pendiente=f^{'}(x)=\dfrac{f(x)}{\Delta x}$", xy=(7,420), xytext=(2, 800),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)
plt.annotate("Raiz real", xy=(1.36881,0), xytext=(-2, 200),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)


plt.text(1.36881,0, r'$\bar{x}$',fontsize=15)
plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x0,fx0, r'$(x_0,f(x_0))$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(3,fx1, r'$(x_1,f(x_1))$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)


plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)

```

## No hay ra√≠ces reales, por ejemplo $f(x)=x^2+2$

```python
def fxOS(x):
    return x**2+2

def gxOS(x):
    return 2*x

x=np.linspace(-3,3)
x0=-1.0
x1=x0-(fxOS(x0)/gxOS(x0))
x2=x1-(fxOS(x1)/gxOS(x1))
x3=x2-(fxOS(x2)/gxOS(x2))
x4=x3-(fxOS(x3)/gxOS(x3))
```

```python
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,fxOS(x))
plt.plot([x0,x1,x2,x3,x4], [0,0,0,0,0], 'o', markersize=12)
plt.plot([x0,x1,x2,x3,x4], [fxOS(x0),fxOS(x1),fxOS(x2),fxOS(x3),fxOS(x4)], 'o',markersize=12)

plt.plot([x0,x0], [0,fxOS(x0)], '--k', markersize=12)
plt.plot([x0,x1], [fxOS(x0),0], '-k', markersize=12)

plt.plot([x1,x1], [0,fxOS(x1)], '--r', markersize=12)
plt.plot([x1,x2], [fxOS(x1),0], '-r', markersize=12)

plt.plot([x2,x2], [0,fxOS(x2)], '--b', markersize=12)
plt.plot([x2,x3], [fxOS(x2),0], '-b', markersize=12)

plt.plot([x3,x3], [0,fxOS(x3)], '--m', markersize=12)
plt.plot([x3,x4], [fxOS(x3),0], '-m', markersize=12)

plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)
plt.text(x3,0, r'$x_3$',fontsize=15)
plt.text(x4,0, r'$x_4$',fontsize=15)

plt.text(x0,fxOS(x0), r'$f(x_0)$',fontsize=15)
plt.text(x1,fxOS(x1), r'$f(x_1$)',fontsize=15)
plt.text(x2,fxOS(x2), r'$f(x_2)$',fontsize=15)
plt.text(x3,fxOS(x3), r'$f(x_3)$',fontsize=15)
plt.text(x4,fxOS(x4), r'$f(x_4)$',fontsize=15)

plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```

## Valor inicial muy lejos de la ra√≠z

```python
def FX(x):
    return -x**3+2*x**2-2
    

def GX(x):
    return -3*x**2+4*x

x=np.linspace(-2,3)
x0=1.0
x1=x0-(FX(x0)/GX(x0))
x2=x1-(FX(x1)/GX(x1))
x3=x2-(FX(x2)/GX(x2))
x4=x3-(FX(x3)/GX(x3))
```

```python
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,FX(x))

plt.plot([x0,x1,x2,x3,x4], [0,0,0,0,0], 'o', markersize=12)
plt.plot([x0,x1,x2,x3,x4], [FX(x0),FX(x1),FX(x2),FX(x3),FX(x4)], 'o',markersize=12)

plt.plot([x0,x0], [0,FX(x0)], '--k', markersize=12)
plt.plot([x0,x1], [FX(x0),0], '-k', markersize=12)

plt.plot([x1,x1], [0,FX(x1)], '--r', markersize=12)
plt.plot([x1,x2], [FX(x1),0], '-r', markersize=12)

plt.plot([x2,x2], [0,FX(x2)], '--b', markersize=12)
plt.plot([x2,x3], [FX(x2),0], '-b', markersize=12)

plt.plot([x3,x3], [0,FX(x3)], '--m', markersize=12)
plt.plot([x3,x4], [FX(x3),0], '-m', markersize=12)

plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)
plt.text(x3,0, r'$x_3$',fontsize=15)
#plt.text(x4,0, r'$x_4$',fontsize=15)

plt.text(x0,FX(x0), r'$f(x_0)$',fontsize=15)
plt.text(x1,FX(x1), r'$f(x_1$)',fontsize=15)
plt.text(x2,FX(x2), r'$f(x_2)$',fontsize=15)
plt.text(x3,FX(x3), r'$f(x_3)$',fontsize=15)
#plt.text(x4,FX(x4), r'$f(x_4)$',fontsize=15)

plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```
